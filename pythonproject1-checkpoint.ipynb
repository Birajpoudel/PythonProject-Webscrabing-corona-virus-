```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.worldometers.info/coronavirus'

soup = requests.get(url)
print(soup)

data = BeautifulSoup(soup.text,'lxml')
print(data)
```

```python
data.table
```

```python
tags=(data.table).find_all('tr')
print(tags)
```

```python
for i in tags:
    print(i.get_text().split())
```
